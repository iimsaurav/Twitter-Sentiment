**Importing dependencies**

```python
import tweepy
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

<br>Importing tweets from twitter through API (Application Interface Protocol) </br>
Signup on *'https://developer.twitter.com/'* for API Authentication Variables.

```python
#My Twitter API Authentication Variables
consumer_key = ' '
consumer_secret = ' '
access_token = ' '
access_token_secret = ' '
```

Imporing Tweets

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)


tweets = api.search('Data', count=50)



data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])

display(data.head(2000))


print(tweets[0].created_at)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()


listy = []

for index, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
  
se = pd.Series(listy)
data['polarity'] = se.values

display(data.head(50))

```
